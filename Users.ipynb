```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to your SQLite database
conn = sqlite3.connect('shiptivity.db')

# Define your SQL query as a multi-line string using triple quotes
query = '''
SELECT
    DATE(login_timestamp, 'unixepoch') AS login_date,
    COUNT(DISTINCT user_id) AS daily_users
FROM
    login_history
GROUP BY
    login_date
'''

# Execute the query and fetch the data
df = pd.read_sql_query(query, conn)

# Create a line chart using Matplotlib
plt.plot(df['login_date'], df['daily_users'])
plt.xlabel('Date')
plt.ylabel('Daily Active Users')
plt.title('Daily Active Users Over Time')
plt.show()
```
